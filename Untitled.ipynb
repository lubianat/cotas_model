Modelo Linear Simples de Desempenho Escolar 

O modelo detalhado aqui é um modelo linear simples para tentar levantar alguns pontos sobre a questão das cotas.

Ele assume que o desempenho de um indivíduo em um exame geral (ex: ENEM) é proporcional a coeficientes que aumentam esse desempenho ao longo do tempo.

O conceito de "desempenho" é uma abstração de resultados diversos. É uma simplificação.

Vamos aos fatores que influenciam esse desempenho:

1 - Tempo

Assumimos, nesse modelo, que quanto mais tempo, maior o desempenho de um aluno. Um aluno do 7º ano do Ensino Fundamental terá um desempenho menor agora do que quando estiver no 3º ano.
Assuminos também que o efeito do tempo no desempenho é linear com o tempo.

desempenho = alfa x tempo + beta

Que é a forma do modelo linear.

Esse coeficiente alfa (o quanto o desempenho muda com o tempo) tem dois componentes:
alfa_total = alfa_1 + alfa_2

* alfa 1, ou a ajuda do meio.
* alfa 2, ou a predisposição individual (facilidade escolar e empenho)

 2 - Alfa 1
 
 O alfa 1 representa o quanto o meio permite o aluno desenvolver seu conhecimento. Condições de vidas precárias, por exemplo, e falta de acesso a recursos educacionais de qualidade, diminuem o alfa 1. Ou seja, ele está bem relacionado ao critério socio-econômico.
 Contudo, outros fatores afetam também esse desempenho. Por exemplo, o racismo estrutural pode contribuir para um menor desempenho. Por exemplo, ao pensarmos em profissões de alto desempenho acadêmico (direito, medicinha e engenharia, por exemplo), pensamos em pessoas brancas.
Isso pode se estender também a outros fatores, como recursos educacionais inacessíveis (por exemplo, cadeirantes e escolas/bibliotecas/museus sem acessibilidade).

3 - Alfa 2

O alfa 2 representa a predisposição individual, que é uma combinação de facilidade com as máterias e empenho.
Para o modelo, assumo (o que julgo coerente) que o alfa_1 (ou seja, as condições do meio) não influenciam essa predisposição.
Ela é considerada algo fixo de cada indivíduo.
De novo, é uma simplificação.

Dessa forma, o desempenho é dado pela equação:
(alfa 1 + alfa 2) x tempo.

4 - O parâmetro Beta
O parâmetro Beta é considerado 0 nesse modelo.
As diferenças de "berço" são embutidas no alfa 1 (coeficiente de influência do meio).
O beta faria sentido num contexto onde, por exemplo, os pais podem "comprar" desempenho.
Por exemplo, em universidades tradicionais americanas, filhos de alunos tem um "bônus" no processo seletivo, que independe do desenvolvimento.

Como estamos pensando no Brasil, isso não será considerado.
Vamos fazer um exemplo simples.

```python

pessoa_a = {}# classe média alta, escola particular, branca. desleixada. 

pessoa_a["alfa_1"] = 0.8
pessoa_a["alfa_2"] = 0.7

pessoa_a["alfa_total"] = pessoa_a["alfa_1"] + pessoa_a["alfa_2"]


pessoa_b = {} # classe média baixa(0.2), escola pública(0.1), parda(0.1). dedicada e com facilidade.

pessoa_b["alfa_1"] = 0.4
pessoa_b["alfa_2"] = 1

pessoa_b["alfa_total"] = pessoa_b["alfa_1"] + pessoa_b["alfa_2"]

pessoa_c = {}  # classe baixa(0.1), escola pública(0.1), parda(0.21. dedicada e com facilidade. 

pessoa_c["alfa_1"] = 0.3
pessoa_c["alfa_2"] = 1.1
pessoa_c["alfa_total"] = pessoa_c["alfa_1"] + pessoa_c["alfa_2"]
```

```python
print(pessoa_a)
print(pessoa_b)
print(pessoa_c)
```

Vemos que o desempenho da pessoa A será mais alto, por que o meio favorece.

Contudo, se igualássemos o meio, ela teria um desempenho menor que as pessoas b e c.

Vamos ver isso em um gráfico.

```python
def calcular_desempenho(pessoa, tempo):
    
    return pessoa["alfa_total"] * tempo
```

```python
anos = range(0,19) 

pessoa_a["desempenho"] = [calcular_desempenho(pessoa_a, ano) for ano in anos ]
pessoa_b["desempenho"] = [calcular_desempenho(pessoa_c, ano) for ano in anos ]
pessoa_c["desempenho"] = [calcular_desempenho(pessoa_c, ano) for ano in anos ]
```

```python
import seaborn as sns; sns.set()
import pandas as pd
import matplotlib.pyplot as plt


df = pd.DataFrame({"A":pessoa_a["desempenho"],
                     "B":pessoa_b["desempenho"],
                  "C":pessoa_c["desempenho"],
                  "tempo":anos})


df_long_format = pd.melt(df, value_vars=["A","B", "C"], id_vars= "tempo", var_name = "pessoa", value_name = "desempenho")

ax = sns.lineplot(x="tempo", y="desempenho", hue="pessoa", data=df_long_format)
```

Vamos supor acgora que as três pessoas entram na universidade.


Vamos supor que a faculdade dá uma educação igual aos indivíduos, e permite várias oportunidades que quase igualam alfa 1 (coeficiente do meio. 
Diferenças estruturais restam, mas a universidade é uma oportunidade.

Vamos supor que Universidade é um período de muito aprendizado novo também.



```python
pessoa_a["alfa_1_na_universidade"] = 0.8 
pessoa_b["alfa_1_na_universidade"] = 0.7 
pessoa_c["alfa_1_na_universidade"] = 0.7


def calcular_desempenho_na_faculdade(pessoa, tempo):
    
    base = calcular_desempenho(pessoa, 18)
    
    # fator multiplicador de 1.5 de desempenho na universidade
    return base + (pessoa["alfa_2"] * tempo + pessoa["alfa_1_na_universidade"] * tempo) * 1.5
```

```python
anos_na_faculdade = range(1,11) 


pessoa_a["desempenho"] += [calcular_desempenho_na_faculdade(pessoa_a, ano) for ano in anos_na_faculdade ]
pessoa_b["desempenho"] += [calcular_desempenho_na_faculdade(pessoa_b, ano) for ano in anos_na_faculdade ]
pessoa_c["desempenho"] += [calcular_desempenho_na_faculdade(pessoa_c, ano) for ano in anos_na_faculdade ]
```

```python
plt.figure(figsize=(20,20))
df = pd.DataFrame({"A":pessoa_a["desempenho"],
                     "B":pessoa_b["desempenho"],
                  "C":pessoa_c["desempenho"],
                  "tempo": list(range(0,29))})


df_long_format = pd.melt(df, value_vars=["A","B", "C"], id_vars= "tempo", var_name = "pessoa", value_name = "desempenho")

ax = sns.lineplot(x="tempo", y="desempenho", hue="pessoa", data=df_long_format)

ax.set_xlim(13,31)
ax.set_ylim(18,60)
```
